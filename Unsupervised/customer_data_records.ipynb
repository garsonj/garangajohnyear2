```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
# Loading the dataset
df = pd.read_csv('customer_data_records.csv')

# Displaying the first few rows and last rows of the dataset
print(df)
print(df.tail())
```

```python
#Select features for clustering
x = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
#Apply KMeans clustering
#use only two clusters
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(x)
```

```python
df['cluster'] = kmeans.labels_
print(df)
```

```python
# Visualizing the clusters
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x ='Annual Income (k$)', y= 'Spending Score (1-100)', hue='cluster')
plt.title('Customer clustering based on Income and Spending')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.show()
```
